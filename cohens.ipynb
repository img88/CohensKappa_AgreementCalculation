```python
from sklearn.metrics import cohen_kappa_score
import pandas as pd
```

```python
df = pd.read_excel('all overlap tweet.xlsx', sheet_name=2)
df.head()
```

```python
df.drop([0], inplace=True)

df.columns = range(df.columns.size)
df.T.reset_index(drop=True).T
df.columns = range(df.shape[1])

df.rename(columns={0: "No", 1: "Tweet", 2: "A1", 3: "A2", 4: "B1", 5: "B2"}, inplace=True)
df.head()
```

```python
mapping = {'Ya': 1, 'Tidak': 0, 'Netral': 1, 'Positif': 2, 'Negatif': 0}
df.replace({'A1': mapping, 'A2': mapping, 'B1': mapping, 'B2': mapping}, inplace=True)
df.head()
```

```python
rater1 = df['A1'].tolist() + df['A2'].tolist()
rater2 = df['B1'].tolist() + df['B2'].tolist()

print(rater1)
print(rater2)
```

```python
cohen_kappa_score(rater1, rater2)
```
